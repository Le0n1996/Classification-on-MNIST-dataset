```python
import numpy as np
import tensorflow
import keras
from keras.datasets import mnist
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
print('Training data shape: ', train_images.shape, train_labels.shape)
print('Testing data shape: ', test_images.shape, test_labels.shape) # let's see what we have
```

We have plenty of observations to build a model. We solve 10-classes classification problem:

```python
Classes = np.unique(train_labels)
nClasses = len(Classes)
print('Total number of outputs: ', nClasses)
print('Output classes: ', Classes)
```

###### Some plotting:

```python
plt.figure(figsize=[10,5])
 
# Display the first image in training data
plt.subplot(121)
plt.imshow(train_images[0,:,:], cmap='gray')
 
# Display the first image in testing data
plt.subplot(122)
plt.imshow(test_images[0,:,:], cmap='gray')
plt.title("Ground Truth: {}".format(test_labels[0]))
```

### Preprocessing:

1. Convert each image matrix (28×28) to an array (28*28 = 784 dimensional) which will be fed to the network as a single feature.

```python
dimData = np.prod(train_images.shape[1:])
train_data = train_images.reshape(train_images.shape[0], dimData)
test_data = test_images.reshape(test_images.shape[0], dimData)
```

2. Convert the data to float and scale it (values between 0 to 1).

```python
train_data = train_data.astype('float32')
test_data = test_data.astype('float32')

# scaling
train_data /= 255
test_data /= 255
# as images in our data are grayscale and the pixel values range from 0 to 255
```

3. Convert the labels from integer to categorical since that is the format required by Keras to perform multiclass classification.
One-hot encoding is used.

```python
# Change the labels from integer to categorical data
train_labels_one_hot = to_categorical(train_labels)
test_labels_one_hot = to_categorical(test_labels)
 
# Display the change for category label using one-hot encoding
print('Original label 0 : ', train_labels[-1])
print('After conversion to categorical ( one-hot ) : ', train_labels_one_hot[-1])
```

### Network

We will be using a network with 2 hidden layers and an output layer with 10 units. The number of units in the hidden layers is kept to be 512. The input to the network is the 784-dimensional array converted from the 28×28 image.

The Sequential model is used for building network. In the Sequential model, we can just stack up layers by adding the desired layer one by one. We use the Dense layer, (fully connected layer) since we are building a feedforward network in which all the neurons from one layer are connected to the neurons in the previous layer. Activation function is ReLu.

The last layer is a softmax layer as it is a multiclass classification problem. For binary classification, we can use sigmoid.

```python
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(dimData,))) # add first hidden layer
model.add(Dense(512, activation='relu')) # add second hidden layer
model.add(Dense(nClasses, activation='softmax')) # add last layer
```

###### Network configuration:

Optimizer is rmsprop. We also specify the loss type, which is categorical cross entropy commonly used for multiclass classification.

We specify the metrics (accuracy in this case) which we want to track during the training process.

Later we try using other optimizers such as adam or SGD.

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

The network is ready to get trained on train data. This is done using the fit() function in Keras. Number of epochs is specified as 20 (the whole dataset will be fed to the network 20 times). Test data will be used for validation.

```python
history = model.fit(train_data, train_labels_one_hot, batch_size=256, epochs=20, verbose=1, 
                   validation_data=(test_data, test_labels_one_hot))
```

#### Evaluation on the test set (using .evaluate() method):

```python
[test_loss, test_acc] = model.evaluate(test_data, test_labels_one_hot)
print("Evaluation result on Test Data: Loss = {}, accuracy = {}".format(test_loss, test_acc))
```

The result on the test set: Loss ~ 0.122, accuracy ~ 0.983

Results look good, but this can be due to overfitting!

### Check for Overfitting:

To do this, we plot the loss and accuracy curves:

```python
#Plot the Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'r',linewidth=3.0)
plt.plot(history.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
 
#Plot the Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['accuracy'],'r',linewidth=3.0)
plt.plot(history.history['val_accuracy'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

As validation loss grows after 5th epoch, this means over-fitting. To fight this, we use regularization - with dropout, we drop half of the neurons, reducing the dependency on the training set by some amount.

```python
model_reg = Sequential()
model_reg.add(Dense(512, activation='relu', input_shape=(dimData,))) # same first layer
model_reg.add(Dropout(0.5)) # drop half of neurons
model_reg.add(Dense(512, activation='relu')) # same second layer
model_reg.add(Dropout(0.5)) # drop half of neurons
model_reg.add(Dense(nClasses, activation='softmax')) # nothing dropped here of course
```

Train it on the train set and check perfomance on the test set:

```python
model_reg.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
history_reg = model_reg.fit(train_data, train_labels_one_hot, batch_size=256, epochs=20, verbose=1, 
                            validation_data=(test_data, test_labels_one_hot))
# training model
```

```python
#Plot the Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history_reg.history['loss'],'r',linewidth=3.0)
plt.plot(history_reg.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves for relu/softmax model with dropout',fontsize=16)
 
#Plot the Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history_reg.history['accuracy'],'r',linewidth=3.0)
plt.plot(history_reg.history['val_accuracy'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves for relu/softmax model with dropout',fontsize=16)
```

As we can see, now validation loss is not increasing that much and training and validation accuracy practically coincide.

Thus, we can say that this model has better generalization capability.
