### MNIST Dataset (in keras)

```python
import numpy as np
import keras
from keras.datasets import mnist
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
import matplotlib.pyplot as plt
```

### Load and explore data

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
print('Training data shape : ', train_images.shape, train_labels.shape)
 
print('Testing data shape : ', test_images.shape, test_labels.shape)
```

So, there are enough observation for exploring and building model

Let's ensure that train set labels are 10

```python
# train labels
classes = np.unique(train_labels)
nClasses = len(classes)
print('Total number of outputs : ', nClasses)
print('Output classes : ', classes)
```

Let's look at the digits plot

```python
plt.figure(figsize=[10,5])
 
# Display the first image in training data
plt.subplot(121)
plt.imshow(train_images[1,:,:], cmap='gray')
plt.title("Ground Truth : {}".format(train_labels[0]))
 
# Display the first image in testing data
plt.subplot(122)
plt.imshow(test_images[5,:,:], cmap='gray')
plt.title("Ground Truth : {}".format(test_labels[0]))
```

### Data preprocessing

Convert each image matrix to array, then we will use it as a single feauture for network

```python
dimData = np.prod(train_images.shape[1:])
train_data = train_images.reshape(train_images.shape[0], dimData)
test_data = test_images.reshape(test_images.shape[0], dimData)
```

Change to float type and scaled data

```python
train_data = train_data.astype('float32')
test_data = test_data.astype('float32')
 
# Scale (between 0 to 1)
train_data /= 255
test_data /= 255
```

Convert the labels from integer to categorical with one-hot since that is the format required by Keras to perform multiclass classification

```python
train_labels_one_hot = to_categorical(train_labels)
test_labels_one_hot = to_categorical(test_labels)
 
# Check the correct form of presentation
print('Original label 0: ', train_labels[0])
print('One-hot representation: ', train_labels_one_hot[0])
```

### Network

Define the network

```python
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(dimData,))) # layer
model.add(Dense(512, activation='relu')) # layer
model.add(Dense(nClasses, activation='softmax')) # activation function
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_data, train_labels_one_hot, batch_size=256, epochs=20, verbose=1, 
                    validation_data=(test_data, test_labels_one_hot))
```

```python
[test_loss, test_acc] = model.evaluate(test_data, test_labels_one_hot)
print("Evaluation result on Test Data : loss = {}, accuracy = {}".format(test_loss, test_acc))
```

```python
# loss curves
plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'r',linewidth=3.0)
plt.plot(history.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves relu/softmax model without dropout',fontsize=16)
 
# accuracy curves
plt.figure(figsize=[8,6])
plt.plot(history.history['accuracy'],'r',linewidth=3.0)
plt.plot(history.history['val_accuracy'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves relu/softmax model without dropout',fontsize=16)
```

### Add regularization

```python
model_reg = Sequential()
model_reg.add(Dense(512, activation='relu', input_shape=(dimData,))) # first layer
model_reg.add(Dropout(0.2)) # set a fraction rate of input units at each update during training time as regularization
model_reg.add(Dense(512, activation='relu')) # second layer
model_reg.add(Dropout(0.2))
model_reg.add(Dense(nClasses, activation='softmax')) # activation function in output layer
```

```python
model_reg.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
history_reg = model_reg.fit(train_data, train_labels_one_hot, batch_size=256, epochs=20, verbose=1, 
                            validation_data=(test_data, test_labels_one_hot))
```

```python
# loss curves
plt.figure(figsize=[8,6])
plt.plot(history_reg.history['loss'],'r',linewidth=3.0)
plt.plot(history_reg.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves relu/softmax model with dropout',fontsize=16)
 
# accuracy curves
plt.figure(figsize=[8,6])
plt.plot(history_reg.history['accuracy'],'r',linewidth=3.0)
plt.plot(history_reg.history['val_accuracy'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves relu/softmax model with dropout',fontsize=16)
```

So, we have very good results for this models in test set!

### New network with more layers, new activation functions, optimizer and loss function

Add more layers with different activation functions

NB! The order of activation function is important!

```python
model_reg_diff_param = Sequential()
model_reg_diff_param.add(Dense(512, activation='relu', input_shape=(dimData,))) # first layer
model_reg_diff_param.add(Dropout(0.7)) # set a fraction rate of input units at each update during training time as regularization
model_reg_diff_param.add(Dense(512, activation='sigmoid')) # second layer
model_reg_diff_param.add(Dropout(0.6))
model_reg_diff_param.add(Dense(512, activation='tanh')) # second layer
model_reg_diff_param.add(Dropout(0.6))
model_reg_diff_param.add(Dense(nClasses, activation='softmax')) # activation function in output layer
```

Change loss function to 'KL divergence' and change optimizer to 'adamax'

```python
model_reg_diff_param.compile(optimizer='adamax', loss='kullback_leibler_divergence', metrics=['accuracy'])
history_reg_diff_param = model_reg_diff_param.fit(train_data, train_labels_one_hot, batch_size=256, epochs=20, verbose=1, 
                                       validation_data=(test_data, test_labels_one_hot))
```

```python
# loss curves
plt.figure(figsize=[8,6])
plt.plot(history_reg_diff_param.history['loss'],'r',linewidth=3.0)
plt.plot(history_reg_diff_param.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves more new settings',fontsize=16)
 
# accuracy curves
plt.figure(figsize=[8,6])
plt.plot(history_reg_diff_param.history['accuracy'],'r',linewidth=3.0)
plt.plot(history_reg_diff_param.history['val_accuracy'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves new settings',fontsize=16)
```

We see that difference between training and validation accuracy is very high. So, we should try to fit models with only 2 or 3 layers, find best optimizer and loss function.
